### One-to-One LSTM for Sequence Prediction

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(len(seq), 1, 1)  #   reshape the 2D sequence into a 3D sequence with 5 samples, 1 time step, and 1 feature.
y = seq.reshape(len(seq), 1)  #  the output as 5 samples with 1 feature.
print("X=",X)
print("y=",y)
```

```python
# define LSTM configuration
n_neurons = length
n_batch = length
n_epoch = 1000
```

```python
# create LSTM
model = Sequential()
# inputs: A 3D tensor with shape [batch, timesteps, feature].
model.add(LSTM(n_neurons, input_shape=(1, 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result:
	print('%.1f' % value)
```

## Many-to-One LSTM for Sequence Prediction (without  TimeDistributed)

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(1, length, 1) # the 2D sequence into a 3D sequence with 1 sample, 5 time steps, and 1 feature.
y = seq.reshape(1, length)  # the output as 1 sample with 5 features.
print("X=",X)
print("y=",y)
```

```python
# define LSTM configuration
n_neurons = length
n_batch = 1
n_epoch = 500
```

```python
# create LSTM
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(length, 1)))
model.add(Dense(length))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result[0,:]:
	print('%.1f' % value)
```

## Many-to-Many LSTM for Sequence Prediction (with TimeDistributed)

<b>There are two key points to remember when using the TimeDistributed wrapper layer:</b>
> 1. The input must be (at least) 3D. This often means that you will need to configure your last LSTM layer prior to your TimeDistributed wrapped Dense layer to return sequences (e.g. set the “return_sequences” argument to “True”).
> 2. The output will be 3D. This means that if your TimeDistributed wrapped Dense layer is your output layer and you are predicting a sequence, you will need to resize your y array into a 3D vector.

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import TimeDistributed
from keras.layers import LSTM
```

```python
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length, 1)
X,y
```

```python
# define LSTM configuration
n_neurons = length
n_batch = 1
n_epoch = 1000
```

```python
# create LSTM
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(length, 1), return_sequences=True))
model.add(TimeDistributed(Dense(1)))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
```

```python
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result[0,:,0]:
	print('%.1f' % value)
```
